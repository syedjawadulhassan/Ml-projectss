```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/indian_cities_weather.csv")
df['date'] = pd.to_datetime(df['date'])

df.head()
```

```python
df.info()
df.describe()
```

```python
city_df = df[df['city'] == "Visakhapatnam"]
city_df.shape
```

```python
plt.figure(figsize=(10,4))
plt.plot(city_df['date'], city_df['avg_temp'])
plt.title("Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Avg Temp")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(city_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```
